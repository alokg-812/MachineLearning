# Linear Regression, Ridge and Lasso

```python
# house pricing dataset
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import seaborn as ss
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = fetch_california_housing()
```

```python
type(df)
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset.head()
```

```python
dataset['Price'] = df.target
```

```python
dataset.head() # the last column 'Price' is a dependent feature which we have inserted in the dataset, rest all are independent feature.
```

```python
## dividing dataset into independent and dependent features
x = dataset.iloc[:,:-1] # skipping the last features and hence getting the independent features
```

```python
y = dataset.iloc[:,-1] # dependent features
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33, random_state = 42)
```

# Linear Regression

```python
# importing all libraries required for linear regression 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
mse = cross_val_score(lin_reg, x_train, y_train, scoring = 'neg_mean_squared_error', cv = 5)
mean_mse = np.mean(mse)
print(mean_mse)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
params = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50, 55, 60, 70, 100]}
ridge_regressor = GridSearchCV(ridge, params, scoring = 'neg_mean_squared_error', cv = 15)
ridge_regressor.fit(x_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
params = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40, 45, 50, 55, 60, 70, 100]}
lasso_regressor = GridSearchCV(lasso, params, scoring = 'neg_mean_squared_error', cv = 5)
lasso_regressor.fit(x_train, y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred = lasso_regressor.predict(x_test)

from sklearn.metrics import r2_score

r2_score = r2_score(y_pred, y_test)
```

```python
print(r2_score)
```

```python
y_pred = ridge_regressor.predict(x_test)

from sklearn.metrics import r2_score

r2_score = r2_score(y_pred, y_test)
print(r2_score)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()

## Independent Features
x = pd.DataFrame(df['data'], columns = df['feature_names'])
```

```python
x.head()
```

```python
## Dependent Features
y = pd.DataFrame(df['target'], columns= ['Target'])
```

```python
y
```

```python
y['Target'].value_counts()
```

```python
## Train Test Split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33, random_state = 3)
```

```python
params = [{'C':[1, 4, 10]}, {'max_iter':[100, 150]}]
```

```python
model1 = LogisticRegression(C = 100, max_iter = 100)
```

```python
model = GridSearchCV(model1, param_grid = params, scoring = 'f1', cv = 5)
```

```python
model.fit(x_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_pred, y_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
